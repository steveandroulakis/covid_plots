```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# grab the data! https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv
# and the UN population data! https://population.un.org/wpp/Download/Archive/Standard/

filename = 'owid-covid-data-18-07-2021.csv'
population_data = 'WPP2019_TotalPopulationBySex.csv'

show_col = 'Total confirmed cases per million (cases per million)'
y_label = 'Total confirmed cases per mil'
y_label2 = 'Total confirmed deaths per mil'


#show_country = 'United States'
show_country = 'United Kingdom'
title = 'Total cases & deaths from COVID-19 per capita: %s' %\
   (show_country)

show_col = 'new_deaths_smoothed'
show_col2 = 'new_cases_smoothed'

title = 'CASES from COVID-19 per capita per day: %s' % show_country
title2 = 'DEATHS from COVID-19 per capita per day: %s' % show_country

df = pd.read_csv(filename)

df['date'] = pd.to_datetime(df['date'], infer_datetime_format=True)
df
```

```python
df = df.sort_values(by=show_col, ascending=False)
df = df[df['date']>'2020-03-01']
df
```

```python
dfpop = pd.read_csv(population_data)

# for per capita calcs
dfpop = dfpop[dfpop['Time']==2020]
dfpop = dfpop[dfpop['Variant']=='No change']
dfpop['Location'] = dfpop['Location'].replace('United States of America', 'United States')
```

```python
show_col_per_capita = '%s_per_mil' % show_col
show_col2_per_capita = '%s_per_mil' % show_col2
df_merge_pop = df.merge(dfpop, left_on='location', right_on='Location')
df_merge_pop

df_merge_pop[show_col_per_capita] = (df_merge_pop[show_col] * (1000/df_merge_pop['PopTotal']))
df_merge_pop[show_col2_per_capita] = (df_merge_pop[show_col2] * (1000/df_merge_pop['PopTotal']))
df_merge_pop
df = df_merge_pop

df
```

```python
show_country_list = [show_country] # todo this is dumb
df = df[df.location.isin(show_country_list)]
df = df.sort_values(by='date', ascending=True)
```

```python
#plot_filename = title.replace(' ', '_').split(':')[0]
plot_filename = "%s_cases_and_deaths" % show_country
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
np.random.seed(12974)


groups = df.groupby('location')

# Plot
fig, (vax, hax) = plt.subplots(1, 2, figsize=(12, 6))
# fig.set_figheight(20)
# fig.set_figwidth(38)
fig.set_figheight(10)
fig.set_figwidth(28)

# Turn on the minor TICKS, which are required for the minor GRID
vax.minorticks_on()
hax.minorticks_on()

#ax.set_ylabel(y_label, fontsize=16)
vax.set_xlabel('Date (source data: ourworldindata.org)', fontsize=16)
vax.set_title(title,
            fontsize=18)

hax.set_xlabel('Date (source data: ourworldindata.org)', fontsize=16)
hax.set_title(title2,
            fontsize=18)

# Customize the major grid
vax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
# Customize the minor grid
vax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

# Customize the major grid
hax.grid(which='major', linestyle='-', linewidth='0.5', color='black')
# Customize the minor grid
hax.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

vax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
hax.margins(0.05) # Optional, just adds 5% padding to the autoscaling
for name, group in groups:
    vax.plot(group['date'], group[show_col2_per_capita],
            marker='o', linestyle='-', label=name,
            linewidth=2, markersize=3)
    
    hax.plot(group['date'], group[show_col_per_capita], color='red',
        marker='o', linestyle='-', label="deaths",
        linewidth=2, markersize=3)
    
plt.savefig('%s.png' % plot_filename)
plt.show()
```
